```python

import requests 
from bs4 import BeautifulSoup 
import shutil
from shutil import make_archive
from zipfile import ZipFile
    
def getdata(url): 
    r = requests.get(url) 
    return r.text 


def getUrls(url,pages):
  urls=[]
  page_urls=range(2,pages)
  for page in page_urls:
    link=url+'/page/'+str(page)
    htmldata = getdata(link) 
    soup = BeautifulSoup(htmldata, 'html.parser') 
    for item in soup.find_all('img'):
      url_path ='http://wallpaperswide.com/'+ item.findParent('a')['href']
      urls.append(url_path)
  return urls

def getImageUrls(urls):
    image_urls=[]
    for url in urls:
      htmldata= getdata(url)
      soup = BeautifulSoup(htmldata, 'html.parser') 
      for item in soup.find_all('a'):
        if item['title'] == "HD 16:9 1920 x 1080 wallpaper for FHD 1080p High Definition or Full HD displays":
          url_path ='http://wallpaperswide.com'+ item['href']
          image_urls.append(url_path)
    return image_urls

def download_images(image_urls):
  for image in  image_urls:
    path ='/content/images/image'+str(image_urls.index(image))+'.jpg'
    response = requests.get(image)
    if response.status_code == 200:
        with open(path, 'wb') as f:
            f.write(response.content)
        print('Image sucessfully Downloaded: ',path)
    else:
        print('Image Couldn\'t be retrieved')
    
```

```python
urls=getUrls("http://wallpaperswide.com/movies-desktop-wallpapers.html",12)
image_urls=getImageUrls(urls)
download_images(image_urls)
```

```python
file = "Images"  # zip file name
directory = "/content/images"
make_archive(file, "zip", directory)  # zipping the directory
print("Contents of the zip file:")
with ZipFile(file+".zip", 'r') as zip:
  zip.printdir()
```
